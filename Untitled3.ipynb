```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
```

```python
airis = load_iris()

df = pd.DataFrame(airis.data, columns=airis.feature_names)

sns.histplot(df['sepal length (cm)'], bins=20, kde=False)
plt.title('Histogram zwykły')
plt.show()

sns.histplot(df['sepal length (cm)'], bins=20, kde=True)
plt.title('Histogram z KDE (zagęszczenie)')
plt.show()
```
